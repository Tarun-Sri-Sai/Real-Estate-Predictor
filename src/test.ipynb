# Loading the model and variables

```python
import pickle as pk
import json
```

```python
model = pk.load(open('../models/linear_regression.sav', 'rb'))
```

```python
load_dict = json.load(open('../cache/input_cache.json', 'r'))
```

# Testing with user input

```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
```

```python
input_dict = {}
```

```python
def update_dict(value, key):
    input_dict[key] = value
    if key in load_dict['encoding_variables']:
      value = load_dict['encodings'][key][input_dict[key]]
    else:
      value = float(input_dict[key])

    print(f'Encoded {input_dict[key]} as {value}')
    input_dict[key] = value
```

```python
def interact_with_dict(key):
    dropdown = widgets.Dropdown(description=f"{key}: ", options=sorted(list(set((load_dict['df_dict'][key]).values()))))
    widgets.interact(update_dict, value=dropdown, key=widgets.fixed(key))
```

```python
for var in load_dict['columns']:
  interact_with_dict(var)
```

```python
output_pred = model.predict(pd.DataFrame(input_dict, index=[0]))
```
